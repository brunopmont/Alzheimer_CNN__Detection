```python
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt
import os

cont = 0

def plot_nifti_slices(nifti_file, slice_indices=None):
    # Carregar a imagem NIfTI
    img = nib.load(nifti_file)
    data = img.get_fdata()

    # Se não forem fornecidos índices de corte, defina alguns padrões
    if slice_indices is None:
        slice_indices = {
            'axial': data.shape[2] // 2,  # Corte axial no meio
            'coronal': data.shape[1] // 2,  # Corte coronal no meio
            'sagital': data.shape[0] // 2   # Corte sagital no meio
        }

    # Configurar a figura
    fig, axs = plt.subplots(1, 3, figsize=(15, 5))

    # Plotar o corte axial
    axs[0].imshow(data[:, :, slice_indices['axial']], cmap='gray')
    axs[0].set_title('Corte Axial')
    axs[0].axis('off')

    # Plotar o corte coronal
    axs[1].imshow(data[:, slice_indices['coronal'], :], cmap='gray')
    axs[1].set_title('Corte Coronal')
    axs[1].axis('off')

    # Plotar o corte sagital
    axs[2].imshow(data[slice_indices['sagital'], :, :], cmap='gray')
    axs[2].set_title('Corte Sagital')
    axs[2].axis('off')

    plt.tight_layout()
    plt.show()

dir = 'train'

category = 'ad'
print('ad')
for item in os.listdir(os.path.join(dir, category)):
    if cont < 3:
        print(item)
        cont += 1
        plot_nifti_slices(os.path.join(dir, category, item))

category = 'mci'
print('mci')
for item in os.listdir(os.path.join(dir, category)):
    if cont < 6:
        print(item)
        cont += 1
        plot_nifti_slices(os.path.join(dir, category, item))

category = 'cn'
print('cn')
for item in os.listdir(os.path.join(dir, category)):
    if cont < 9:
        cont += 1
        print(item)
        plot_nifti_slices(os.path.join(dir, category, item))
```
